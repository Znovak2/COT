```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

movies_file = 'data/movielens/movies.dat'
ratings_file = 'data/movielens/ratings.dat'
users_file = 'data/movielens/users.dat'

movies_df = pd.read_csv(movies_file, delimiter='::', engine='python', names=["MovieID","Title","Genres"])
ratings_df = pd.read_csv(ratings_file, delimiter='::', engine='python', names=["UserID","MovieID","Rating","Seconds"])
users_df = pd.read_csv(users_file, delimiter='::',engine='python', names=["UserID","Gender","Age","Occupation","Zip-code"])
```

#### Number 1
Transform Titles lowercase to uppercase and uppercase to lowercase.

#### Number 2
Replace movie id with any number between 99 to 999 (use conditions as needed)

The below cell contains my test framework to create the convert function which will eventually convert MovieID values into quandrants. These quadrants do not neccesarily compute any significance unless there is a relation in the IMDB's provisioning process of MovieIDs. 

```python
import random

def testconvert(MovieID):
    if 1 <= MovieID < 549:
        return 1
    elif 549 <= MovieID <= 999:
        return 2
    else:
        return 0

test = np.random.randint(99,999)
test_convert = testconvert(test)
test_convert
```

In the Output above, I can confirm my framework tested succesfully. Now, I will add more conditions specific to the instructed range. I use the .max() method to find the largest MovieID value since I know the smallest is 1.

```python
ratings_df1 = ratings_df.copy()
print("Maximum MovieID: ",ratings_df1['MovieID'].max())

def convert(MovieID):
    if 1 <= MovieID < 988:
        return 100
    elif 989 <= MovieID < 1976:
        return 200
    elif 1977 <= MovieID < 2964:
        return 300
    elif 2965 <= MovieID <= 3952:
        return 400
    else:
        return 0

test = np.random.randint(1,3952)
print("Test MovieID: ", test)
verify_convert = convert(test)
print("Quadrant Value: ",verify_convert)
```

Now, time to replace the MovieID values in ratings_df by using the .apply() method and my convert function defined above.

```python
ratings_df2 = ratings_df.copy()
ratings_df2['MovieID'] = ratings_df2['MovieID'].apply(convert)
ratings_df2.sample(n=10)
```

#### Number 5
Split Genre by the prefix 

First, I create a new data frame that is generated from the splitting of the Genres column. By setting expand to True, values will be split into their own column and create a new column until there are no more splits occuring.

```python
movies_df4 = movies_df.copy()

genre_column = movies_df4['Genres'].str.split('|',expand=True)
genre_column.columns = ['Primary Genre','Secondary Genre','Third Genre','Fourth Genre','Fifth Genre','Sixth Genre']
genre_column
```

Next, I combine the two using pd.concat since Append() will be depracated soon. By specifying axis=1, the genre_column dataframe was added to the right/horizontally of the already existing movies_df copy (movies_df4). With the resulting newdf dataframe, I used the drop method to remove the combined Genre column as it was now redundant.

```python
newdf = pd.concat([movies_df4,genre_column],axis=1)
newdf = newdf.drop(newdf.columns[2], axis=1)
newdf
```

###### Number 1
Provide a regular expression that may be used to extract complete email addresses. Use my email id ramya.akula@ucf.edu

```python
import re

text = "Use my email id ramya.akula@ucf.edu"
text_list = text.split()

regex1 = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9]+\.[A-Za-z]{2,}\b'


for name in text_list:
        result=re.search(regex1,name)
        if result:
            print(result)
```
