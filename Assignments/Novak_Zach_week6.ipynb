### 3. Read the CSV data (TATA MOTORS in Modules) into a dataframe.

```python
# read stock data from csv file
stock_data = pd.read_csv('data/Tata_Motors_Ltd._historical_data.csv')
# preview the data
stock_data.head()
```

```python
# add year and month columns to the dataframe
stock_data['Year'] = pd.DatetimeIndex(stock_data['Date']).year
stock_data['Month'] = pd.DatetimeIndex(stock_data['Date']).month
# preview the data
stock_data.head()
```

## Financial Charts

### 4. Write down the code for the Candlestick Chart Below.

```python
# create a candlestick chart using plotly
fig = go.Figure(data=[go.Candlestick(
    x=stock_data['Date'],
    open=stock_data['Open'],
    high=stock_data['High'],
    low=stock_data['Low'],
    close=stock_data['Close']
)])

# add title and axis labels
fig.update_layout(title='Tata Motors Ltd. Stock Price',
                  xaxis_title='Date',
                  yaxis_title='Stock Price')

# show the chart
fig.show()
```

### 6. Write down the code for the Area Chart below.

```python
# create a line chart using plotly
fig = go.Figure()
# add a trace for the closing price
fig.add_trace(go.Scatter(
    x=stock_data['Date'], # x-axis data
    y=stock_data['Close'], # y-axis data
    fill='tozeroy',  # fill area under the line
    mode='lines', # line mode
    name='Close Price'  # name of the trace
))
# add title and axis labels
fig.update_layout(title="Area Chart of Tata Motors' Closing Price", xaxis_title="Date", yaxis_title="Price", showlegend=True)
# show the chart
fig.show()
```
