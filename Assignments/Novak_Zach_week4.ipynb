## Part 1: 
Read Total profit for each month and demonstrate it using a line plot with the following Style properties:
- Line Style Dashed and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Profit in Dollars
- Add a circle marker
- Line width should be 3

```python
# import modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# read and preview the data
sales_df = pd.read_csv('data/company_sales_data.csv')
sales_df.head()
```

```python
# setup plot variables
x = sales_df['month_number']
y = sales_df['total_profit']

# plot the data
plt.plot(x, y, label='Profit data of last year', color='r', marker='o', markerfacecolor='k', linestyle='--', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Profit in Dollars')
plt.legend(loc='lower right')
plt.title('Company Profit for Trailing 12 months')
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Part 2:
Read toothpaste sales data of each month and show it using a scatter plot, adding a grid in the plotline.
- Gridline style should be "--"

```python
# setup plot variables
x = sales_df['month_number']
y = sales_df['toothpaste']

# plot the data
plt.scatter(x, y, label='Toothpaste Sales', color='b', marker='o')
plt.grid(True, linewidth=1, linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Number of units Sold')
plt.legend(loc='upper left')
plt.title('Tooth paste Sales Trailing 12 Months')
plt.xticks(x)
plt.yticks([4500, 5000, 5500, 6000, 6500, 7000, 7500, 8000])
plt.show()
```

## Part 3:
Calculate total sale data for last year for each product and show it using a Pie chart.

```python
# setup variables
skus = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# calculate total units sold across all products
total_units_sold = sales_df[skus].sum().sum()

# calculate percentage of each product units sold
sku_units_sold = sales_df[skus].sum()
sku_sales_percent = (sku_units_sold / total_units_sold) * 100 # percentage of each product units making up total sales


# setup plot variables like capitalize the labels
labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

# plot the data
plt.pie(sku_units_sold, labels=labels, colors=colors, autopct='%1.1f%%') # autopct to show percentage
plt.legend(loc='lower right')
plt.title('Sales Data')
plt.show()
```

## Part 4:
Find the count of total car for each car making company.

```python
# read and preview the data
car_data = pd.read_csv('data/automobile_data.csv')

# print the head, description, info, and shape of the data
print("Head of the DataFrame:")
print(car_data.head(1))

print("\nDescription of the DataFrame:")
print(car_data.describe())

print("\nInfo of the DataFrame:")
car_data.info()

print("\nShape of the DataFrame:")
print(car_data.shape)
```

```python
# convert dtypes to more applicable types
car_data = car_data.astype({
    'company': 'category', # category saves memory when there are few unique values
    'body-style': 'category', # category saves memory when there are few unique values
    'engine-type': 'category', # category saves memory when there are few unique values
    'num-of-cylinders': 'category',  # category saves memory when there are few unique values
    'wheel-base': 'float32', # float32 saves memory compared to float64
    'length': 'float32', # float32 saves memory compared to float64
    'horsepower': 'int16', # int16 saves memory compared to int64
    'average-mileage': 'int16', # int16 saves memory compared to int64
    'price': 'float32' # float32 saves memory compared to float64
})

print(car_data.dtypes)
```

```python
# find out how many cars are manufactured by each company
# iterate over the unique car companies in the dataset
for company in car_data['company'].unique():
    # count the number of cars (rows) manufactured by each company
    total_car_skus = car_data[car_data['company'] == company].shape[0] # shape[0] gives the number of rows
    print(f"{company.capitalize()} makes {total_car_skus} car(s).")
```

## Part 5:
Delete the second column from a given array and insert the following new column in its place:
- sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]])
- newColumn = np.array([[10,10,10]])

```python
# setup variables
sampleArray = np.array([[34,43,73],[82,22,12],[53,94,66]])
newColumn = np.array([[10,10,10]])

# print the original array
print("Original sampleArray:\n", sampleArray)

# delete the second column
sampleArray = np.delete(sampleArray, 1, axis=1)
print("\nsampleArray after deleting the second column:\n", sampleArray)

# insert the new column at the second position
sampleArray = np.insert(sampleArray, 1, newColumn, axis=1)
print("\nsampleArray after inserting the new column:\n", sampleArray)
```
