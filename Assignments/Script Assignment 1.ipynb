### 11. Visualization the data (Use: Stacked bar chart or any of your choice) - Bonus - 5 point

```python
from matplotlib import pyplot as plt

hungary = hungary.copy()

hungary['TIME'] = hungary['TIME'].astype(int)
hungary['Value'] = hungary['Value'].astype(float)


hg_ordered = hungary.sort_values(by='TIME')

year = hg_ordered['TIME']
values = hg_ordered['Value']

fig = plt.figure(figsize = (10,5))

plt.bar(year,values, color ='blue', width = 0.4)

plt.xlabel("Year")
plt.ylabel("Education allocation (in percents)")
plt.title("Hungary's education allocation over the years")
plt.show()
```
